### For now I am not sure for the 'binnedInc' column, so I am just gonna drop it. If afterwards we felt the need we may use it.
### Also we perform one hot encoding on "Geography" column as it is nominal categorical data.

```python
df.drop("binnedInc", axis = 1, inplace = True)
df1 = pd.get_dummies(df["Geography"])
```

### Now we have to deal with the Null values 

```python
import seaborn as sns
sns.heatmap(df.isnull(), yticklabels = False, cbar = False)
```

```python
df.info()
```

### The purpose of seeing info is to observe that if there is any null values or not and also to see that whether all our columns are converted into datatype that can be used in model fitting.

```python
lis_null = ['PctSomeCol18_24', 'PctEmployed16_Over', 'PctPrivateCoverageAlone']
for i in lis_null:
    df[i] = df[i].fillna(df[i].mean())
```

## Data Preprocessing
### We plot a correlation matrix to observe what features are highly corelated to each other and what features are not correlated with the death rates. The purpose is that if some features are very little correlated with the dependent variable, we can drop that column as it won't contribute much in test set prediction.

```python

corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

### Here are some interesting observtions from the correlation matrix.
### 1. Avganncount and AvgDeathperYear are highly correlated. So one can be droped. However 57 is not a very large number for the parameters, so we keep both.
### 2. AvgDeathperYear and population of county are also highly correlated(0.977634).
### 3. PovertyPercent and medIncome are highly -ve correlated, which means if one value goes up other goes down.

### We drop the column which have less than 0.04 correlation value with DeathRate column or our dependent variable

```python
corr["TARGET_deathRate"]
lis_col = list(corr.columns)
for i in range(len(corr["TARGET_deathRate"])):
    if abs(corr["TARGET_deathRate"][i]) < 0.04:
        df.drop(lis_col[i], axis = 1, inplace = True)
```

```python
df
```

### We significantly reduced our columns from 83 to 57. It would be easie for our model to train the dataset.

```python
train = df.copy()
train.drop("TARGET_deathRate", axis = 1, inplace = True)
Y = df["TARGET_deathRate"]
```

```python
train.columns
```

### By seeing the correlation matrix, one other interesting thing to notice is that among white, black, asian and other race only Black column is positively correlated with the deathrate column. It means that the death rate may be more prone to areas where black population is increased.

```python
plt.subplot(2, 2, 1)
plt.scatter(df["PctWhite"], df["TARGET_deathRate"])

plt.subplot(2, 2, 2)
plt.scatter(df["PctBlack"], df["TARGET_deathRate"])

plt.subplot(2, 2, 3)
plt.scatter(df["PctAsian"], df["TARGET_deathRate"])

plt.subplot(2, 2, 4)
plt.scatter(df["PctOtherRace"], df["TARGET_deathRate"])
```

### Normalizing the data into the range of (0, 1) both inclusive.

```python
for i in train.columns:
    train[i] = (train[i] - min(train[i]))/(max(train[i] - min(train[i])))   
train
```

```python
Y = (Y - min(Y))/(max(Y)- min(Y))
Y.hist()
```

### Splitting the training set into Train and Cross Validation set. Cross Validation set is used to check if our model overfits or not.

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_cv, Y_train, Y_cv = tts(train, Y, test_size = 0.2, random_state = 0)
```

# MODEL FITTING

### We are going to use two models , i) Ridge Regression technique which is basically linear regression with Regularization and ii) GradientBoostingRegressor.

## MODEL-1( RIDGE REGRESSION)

```python
from sklearn.linear_model import Ridge

model1 = Ridge(random_state = 0)
model1.fit(X_train, Y_train)
pred1_train = model1.predict(X_train)
pred1_cv = model1.predict(X_cv)
plt.scatter(pred1_cv, Y_cv)
```

```python
loss1_train = (pred1_train-Y_train)**2
J_train = (1/(2*len(Y_train)))*(loss1_train.sum())
J_train
```

```python
loss1_cv = (pred1_cv-Y_cv)**2
J_cv = (1/(2*len(Y_cv)))*(loss1_cv.sum())
J_cv 
```

```python
#Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid1 = {'alpha': [0.01, 0.03, 0.1, 0.3, 1.0, 3.0, 10, 30, 100, 300, 1000]}
m1 = GridSearchCV(estimator = model1, param_grid = param_grid1, 
                          cv = 3, n_jobs = -1, verbose = 2)
m1.fit(X_train, Y_train)
```

```python
m1.best_params_
```

```python
pred1_train = m1.predict(X_train)
pred1_cv = m1.predict(X_cv)
plt.scatter(pred1_cv, Y_cv)
```

```python
from sklearn.metrics import mean_squared_error
print("The RMSE error in predicting the training examples is:",mean_squared_error(pred1_train, Y_train))
print("The RMSE error in predicting the cross-validation examples is:",mean_squared_error(pred1_cv, Y_cv))
```

## MODEL-2( GBR)

```python
from sklearn.ensemble import GradientBoostingRegressor as gbr
reg = gbr(random_state = 0)

model2 = reg.fit(X_train, Y_train)
pred2_train = model2.predict(X_train)
pred2_cv = model2.predict(X_cv)
print("The RMSE error in predicting the training examples is:",mean_squared_error(pred2_train, Y_train))
print("The RMSE error in predicting the cross-validation examples is:",mean_squared_error(pred2_cv, Y_cv))
```

```python
# PERFORMING THE HYPER-PARAMETER TUNING ON GBR MODEL
param_grid2 = {
    'learning_rate': [0.03, 0.1, 0.3, 1.0, 10],
    'max_depth': [1, 2, 3],
    'n_estimators': [100, 200, 300]
}
m2 = GridSearchCV(estimator = reg, param_grid = param_grid2, 
                          cv = 3, n_jobs = -1, verbose = 2)
m2.fit(X_train, Y_train)
```

```python
pred2_train = m2.predict(X_train)
pred2_cv = m2.predict(X_cv)
print("The RMSE error in predicting the training examples is:",mean_squared_error(pred2_train, Y_train))
print("The RMSE error in predicting the cross-validation examples is:",mean_squared_error(pred2_cv, Y_cv))
```

### We see that for our GradientBoostingRegressor model after hyperparameter tuning we get J_train error of order 0.00019 and for J_cv error is of 0.00248.
### It is a pretty good fit and the accuracy cn also be increased after doing some Ensembling Techniques.Here I am using Average method.

```python
pred_train = (pred1_train + pred2_train)/2
pred_cv = (pred1_cv + pred2_cv)/2
```

```python
print("The RMSE error in predicting the training examples is:",mean_squared_error(pred_train, Y_train))
print("The RMSE error in predicting the cross-validation examples is:",mean_squared_error(pred_cv, Y_cv))
```
